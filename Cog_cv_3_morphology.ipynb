```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('carnum.jpeg')
img.shape
```

```python
img = cv2.resize(img, (300, 225))
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# rectangular kernel -> BLACKHAT ops-> find dark regions on light background
# WHITEHAT ops-> find light regions on dark background
```

```python
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13,5))
blackhat_map = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT ,rectKernel)
```

```python
topHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, rectKernel)
```

```python
cv2_imshow(blackhat_map)
```

```python
cv2_imshow(topHat)
```

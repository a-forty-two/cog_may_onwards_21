```python
import cv2
import imutils
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('1.jpg')
cv2_imshow(img)
```

```python
(h,w,d) = img.shape
print(h)
print(w)
print(d)
```

```python
(B,G,R) = img[200,500]
print(B)
print(G)
print(R)
```

```python
slice = img[120:220 , 300:-200]
cv2_imshow(slice)
```

```python
cv2_imshow(resized)
```
